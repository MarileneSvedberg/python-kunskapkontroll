### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

**Answer:**
There are two ways:
1) The pyplot (state-based approach): A simple and convenient way of creating plots is using the pyplot. It is based on the state of the current figure and axes, wich means we don't explicity create figure or axes objects. We use functions like plt.plot(), plt.xlabel(), plt.ylabel(), etc. 

2) The Object-Oriented approach: This approach is more flexible and recommended for creating more complex plots. Its envolves creating Figure and Axes objects explicitly, and then using methods of theses objects to plot and customize the figure. This method provides better control over plot especially for multiple subplots or complex layouts

(Explicit gränssnitt - ett Matplotlib-diagram ritas upp på ett objekt som kallas figure. En Figure håller koll på allt som diagrammet består av - linjer, prickar, staplar, titlar, etiketter(labels). Själv diagrammen ritas på objekt som kallas Axes. En figure kan bestå av en eller flera Axes. När vi använder det explicita gränsnitt skapar vi själva en Figure och så många axes som vi behöver. Det vanligastes sättet är att använda metoden subplots(). Den returnerar en tuple med ett Figure-objekt och så många Axes-objekt som vi anger)

   Den officiella rekommendationen från Matploblib är att man använder det explicita, objektorienterade grännsnitt.

    

```python
# Exempel för The Pyplot(State-based approach)
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
#print(x)
y = np.sin(x)

plt.plot(x, y)  # Simple plot using pyplot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine Wave')
plt.show
```

```python
#Exempel för Object-Oriented-approach (Den rekommenderade approach att användas)

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()  # Create a figure and axes explicitly
ax.plot(x, y)             # Plot on the axes object
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Sine Wave')
plt.show()
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

**Answer**
A figure is a window or a container that holds everything in the plot, including axes, titles, labels, and other elements. A figure can contain multiple axes (subplots)

Example:
fig = plt.figure()

Axes are the part of the figure where data is plotted. They are the are that contains the x-axis and y-axis, and the plot itself. A figure can have multiples  axes (for multiples subplots). We create axes with fig.add_axes() or by using plt.subplots().
Example:
fig, ax = plt.subplots ()

Axis: An axis refers specifically to the x-axis or y-axis in the plot. We can modify the properties of the axis, such as its labels, limits, tick marks, etc.

Artis: An artist is any object that can be drawn on the figure, including lines, text, patches (like rectangles or circles), images, labels, and so on.
Essentially, artists are the visual elements that make up the plot, such as the data line created by plot(), or the title created by set_title().

In summary:
Figure: The entire window or canvas containing the plot.
Axes: The area within the figure where data is plotted.
Axis: The individual x or y axis controlling the data range and ticks.
Artist: Any visual element in the plot, such as lines, labels, or markers.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function y = x^2
x = np.linspace(-4, 4, 100)  # Create an array of 100 points from -4 to 4
y = x**2  # Calculate y = x^2

# Object-Oriented approach:
fig, ax = plt.subplots()  # Create figure and axes
ax.plot(x, y)  # Plot the function y = x^2
ax.set_title("Plot of y = x^2")  # Set title
ax.set_xlabel("x")  # Set x-axis label
ax.set_ylabel("y")  # Set y-axis label

# Show the plot
plt.show()
```

```python
# Pyplot approach:
plt.plot(x, y)  # Plot the function y = x^2
plt.title("Plot of y = x^2")  # Set title
plt.xlabel("x")  # Set x-axis label
plt.ylabel("y")  # Set y-axis label

# Show the plot
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2


# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 6))

# Scatter plot
ax1.scatter(random_data_x, random_data_y, alpha = 0.5, color ="green")
ax1.set_title("Scatter Plot of Random Data")
ax1.set_xlabel("X-axis")
ax1.set_ylabel("Y-axis")

# Bar plot
ax2.scatter(names, values, color ="green")
ax2.set_title("Fruit Data")
ax2.set_xlabel("Fruit")
ax2.set_ylabel("Quantity")
```
